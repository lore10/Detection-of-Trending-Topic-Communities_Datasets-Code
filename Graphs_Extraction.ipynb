# RBC: Retweeting Graphs per trending topic

368 trending topics done for Retweeting Graph RBC method

```python
for topic in tt_string:
    G=nx.DiGraph()
    
    cr=[(u[0],u[1],u[2],u[3],u[4]) for u in creators if u[2]==topic]
    di=[(u[0],u[1],u[2],u[3],u[4]) for u in distributors if u[2]==topic]
    
    cr_ids=[u[0] for u in cr]
    di_ids=[u[0] for u in di]
    repetidos=list(set(di_ids).intersection(set(cr_ids)))  #eliminate the distributors who were found as CREATORS
    
    di=[(u[0],u[1],u[2],u[3],u[4]) for u in di if u[0] not in repetidos]
    
    retweeting_pairs=[]
    for creator in cr:
        for distributor in di:
            if distributor[4] == creator[4]: #if the distributor retweeted the tweet t, then join the distributor with the creator
                retweeting_pairs.append((distributor[0], creator[0]))
    retweeting_pairs=list(set(retweeting_pairs))
    save_json('Retweeting_graphs368/'+topic, retweeting_pairs)
              
    nodes = list(set([n1 for n1, n2 in retweeting_pairs] + [n2 for n1, n2 in retweeting_pairs]))
    G.add_nodes_from(nodes)
    G.add_edges_from(retweeting_pairs)
    nameFile='Retweeting_graphs368/'+topic+'.gexf'
    nx.write_gexf(G,nameFile)
    G.clear()
```

# Following Graphs per trending topic for CREATORS

Create following graph among Creators

```python
for topic in tt_string:
    G=nx.DiGraph()
    nameFile='FollowersCreators_perTrendingTopic/'+topic
    topic_folowers= load_json(nameFile)
    
    cr=[(u[0],u[1],u[2],u[3],u[4]) for u in creators if u[2]==topic]
    cr_ids=[u[0] for u in cr]
    
    following_pairs_user=[]
    following_pairs=[]
    
    for user in topic_folowers:
        following_pairs_user=[(inf, user['user_id']) for inf in cr_ids if inf in user['followers']]
        following_pairs+=following_pairs_user

    following_pairs=list(set(following_pairs))
    save_json('Following_graphs_Creators368/'+topic, following_pairs)
              
    nodes = list(set([n1 for n1, n2 in following_pairs] + [n2 for n1, n2 in following_pairs]))
    G.add_nodes_from(nodes)
    G.add_edges_from(following_pairs)
    nameFile='Following_graphs_Creators368/'+topic+'.gexf'
    nx.write_gexf(G,nameFile)
    G.clear()
                
```

# Following graphs for DISTRIBUTORS

Create following graph among Distributors

```python
users_ids=load_json('unique_users_Creators_Distrib_perTopic368')
```

```python
for topic in tt_string:
    G=nx.DiGraph()
    nameFile='FollowersDistributors_perTrendingTopic/'+topic
    topic_folowers= load_json(nameFile)
    
    di=[u['distributors'] for u in users_ids if u['topic']==topic][0]
    
    following_pairs_user=[]
    following_pairs=[]
    
    for user in topic_folowers:
        following_pairs_user=[(susc, user['user_id']) for susc in di if susc in user['followers']]
        following_pairs+=following_pairs_user

    following_pairs=list(set(following_pairs))
    save_json('Following_graphs_Distributors368/'+topic, following_pairs)
              
    nodes = list(set([n1 for n1, n2 in following_pairs] + [n2 for n1, n2 in following_pairs]))
    G.add_nodes_from(nodes)
    G.add_edges_from(following_pairs)
    nameFile='Following_graphs_Distributors368/'+topic+'.gexf'
    nx.write_gexf(G,nameFile)
    G.clear()
```
